## Hent data for hvert ord

Den her tar et par minutter. Tid til å lage en kopp te!

```python
df = []
for qv in qvinder:
    #print(qv)
    x = nb.unigram(qv, period=(1830, 1880), media='bok')
    if not x.empty:
        x.columns = [qv]
        df.append(x)
```

### Summer opp totalen over ord, og for år (nedenfor) 

```python
tot = pd.concat(df, axis='columns').fillna(0)
```

```python
tot
```

Se på totalsummer og sett grenser for øver og nedre. Spennet er såpass stort at mye domineres av de ordene med størst frekvens

```python
totsum = tot.sum()
```

Delsummer og studier av dem. Her er det snakk om totalantall over hele perioden. Delmengder er knyttet til et parameter, bare endre på tallet for `totsum`. Maxen her er på flere tusen. Du kan også endre på selve relasjonen, og skrive `totsum` > 300, for eksempel. Trykk på code_toggle-knappen for å se koden.

```python
subsum = totsum[totsum < 5]
subsum.plot(kind='barh', figsize = (15, int(len(subsum)*0.4)));
```

## Summasjoner pr. år - her totalfrekvens:

```python
annosum = tot.sum(axis=1)
```

```python
annosum.plot(figsize=(10, 5), lw=4, title="Totalfrekvenser pr. år");
```

## Relativfrekvens, relativ til punktueringer

```python
punktueringer = ". , og i".split()
df_punkt = pd.concat([nb.frame(nb.unigram(p, period=(1830, 1900), media='bok'), p) for p in punktueringer], axis='columns')
```

```python
df_punkt_sum = nb.frame(df_punkt.sum(axis=1), 'punkt')
```

```python
df_punkt_sum.plot(figsize = (10,5), lw=4, title = "Totalfrekvens for punktuering");
```

## Plot over kvinneord relativ til punktueringer. 

Omtrent like bra som (om ikke bedre enn) relativfrekvens sammenlignet med totalen. Punktueringene har et stabilt prosentvis antall av totalen. 

```python
(annosum/df_punkt_sum['punkt']).rolling(window=5).mean().plot(figsize=(10, 5), lw=3, title="Kvinneord pr. punktueringer");
```

# Samme trend om pronomene tas ut

Tallene blir bare litt mindre, ellers ser grafen ganske lik ut som over, men merk at økningen ikke er like stor. Når pronomene er borte går den fra 0.0025 til 0.0055, en dobling omtrent, mens med pronomen går den dra 0.006 til 0.018, en tredobling ca. 

```python
anno_uten_hun = tot[list(set(tot.columns) - set(['Hun', 'hun','hende', 'Hende', 'hendes', 'Hendes']))]
```

```python
(anno_uten_hun.sum(axis=1)/df_punkt_sum['punkt']).rolling(window=5).mean().plot(figsize=(10, 5), lw=3, title= "Kvinner uten pronomen pr. punktuering");
```
